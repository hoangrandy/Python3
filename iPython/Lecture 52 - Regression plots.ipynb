```python
# Normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import stats library from numpy
from scipy import stats

# Import plotting modules and libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#sns.lmplot for regression
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips, 
          scatter_kws={'marker':'o','color':'red'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips, order =4,
          scatter_kws={'marker':'o','color':'red'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
# For no regression plot
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
# Making a new column that will calculate the tip percentage
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
# Jitter wiki
url = 'https://en.wikipedia.org/wiki/Jitter'
```

```python
# Makes it easier to read
sns.lmplot('size','tip_pect',tips,x_jitter=.1)
```

```python
# This shows the tendency for each bin
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
# hue wil automatically make hues for the sex columns
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
url = 'http://en.wikipedia.org/wiki/Local_regression'
# Let see what local regression can do for us
sns.lmplot('total_bill','tip_pect',tips, lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill', 'tip_pect',tips)
```

```python
# Let's create two subpltos
fig, (axis1,axis2) = plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(tips['tip_pect'],tips['size'],color='red',ax=axis2)
```
