```python
# Plotting 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# allows us to see the plots in ipython 
%matplotlib inline
```

```python
dataset1 = randn(100)
```

```python
plt.hist(dataset1)
# plt.hist is defaulted to 10 equal bins
```

```python
# Changing colors
dataset2 = randn(80)

plt.hist(dataset2, color ='red')
```

```python
# See both histograms in the same plot
# you have to use density to normalize the different # of plots
# To control transparency use alpha
plt.hist(dataset1, density = True, color = 'red', alpha=0.5, bins=20)
plt.hist(dataset2, density = True, color = 'blue', alpha = 0.5, bins =20)
```
