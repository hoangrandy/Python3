```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

#import the stats library from numpy
from scipy import stats

#These are the plotting modules and librarys
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot([data1, data2])
```

```python
sns.boxplot([data1,data2], whis=np.inf)
```

```python
#normal Dist
data1 = stats.norm(0,5).rvs(100)

# Two gama dis. Concatenated together 
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1*stats.gamma(5).rvs(50)])
#.inf = infinity

#Box plot both data1 and data2
sns.boxplot([data1,data2], whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.01)
```

```python
# You show individual points, each line are data points
sns.violinplot(data1, inner = 'stick')
```
